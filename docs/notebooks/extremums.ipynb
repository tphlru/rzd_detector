```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

y = np.array(values)
x = np.array(range(len(y)))

# Находим пики (максимумы) и впадины (минимумы)
extremums = find_peaks(y, distance=40, prominence=55)[0].tolist()
extremums += find_peaks(-y, distance=40, prominence=60)[0].tolist()

# Если нам попались супер-неправильные данные и фильтры всё уничтожили
if len(extremums) == 0:
    extremums = find_peaks(y, distance=20, prominence=None)[0].tolist()
    extremums += find_peaks(-y, distance=20, prominence=None)[0].tolist()

# Вручную ищем на концах, так как find_peaks этого сделать не может

# В начале игнорируем (закомментированно)
# if np.sign(np.diff(y))[0] < 0 or np.sign(np.diff(-y))[0] < 0:
#     extremums.append(0)
if np.sign(np.diff(y))[-1] > 0 or np.sign(np.diff(-y))[-1] > 0:
    extremums.append(len(y) - 1)

extremums.sort() # Отсортируем для правильного отображения линий
extremums_values = y[extremums].tolist()

plt.title("Дыхание")
plt.plot(x, y, label='Data')
plt.plot(x[extremums], extremums_values, color="red", label='Экстремумы')
plt.legend()
plt.show()
```
